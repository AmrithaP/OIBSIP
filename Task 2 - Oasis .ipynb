Unemployment is measured by the unemployment rate which is the number of people
who are unemployed as a percentage of the total labour force. We have seen a sharp
increase in the unemployment rate during Covid-19, so analyzing the unemployment rate
can be a good data science project.

```python
# Loading the Dataset 2
Unemployment_data2 = pd.read_csv("Unemployment_Rate_upto_11_2020.csv")
Unemployment_data2
```

```python
# convert to appropriate Data types for analysis - for Date
Unemployment_data2[" Date"]= Unemployment_data2[" Date"].astype('datetime64[ns]')
Unemployment_data2['Year']=Unemployment_data2[" Date"].dt.year
Unemployment_data2['Month']=Unemployment_data2[" Date"].dt.month
Unemployment_data2=Unemployment_data2.drop([" Date"],axis=1)
```

```python
# By Region.1
count_region = Unemployment_data2['Region.1'].value_counts()
count_region
```

```python
count_region.plot(kind='pie',startangle = 90)

plt.title("Region wise Unemployment")
plt.show()
# North region is seen to have maximum unemployment
```

```python
# By Region
count_region = Unemployment_data2['Region'].value_counts()
count_region
```

```python
# 
plt.figure(figsize=(18,10))
sns.barplot(x = 'Region', y = ' Estimated Unemployment Rate (%)', data = Unemployment_data2)
plt.ylabel(' Estimated Unemployment Rate', fontsize=12)
plt.xlabel('Region/State', fontsize=12)
plt.title("Unemployment Rate in each state", fontsize=15)
plt.xticks(rotation='vertical')
plt.show()
#Haryana is seen to have max unemplyment
```

```python
plt.figure(figsize=(18,10))
sns.barplot(x = 'Region', y = ' Estimated Employed', data = Unemployment_data2)
plt.ylabel(' rate', fontsize=12)
plt.xlabel('Region', fontsize=12)
plt.title("Estimated Employed individuals in Region", fontsize=15)
plt.xticks(rotation='vertical')
plt.show()
# Uttar Pradesh has max employed individuals
```

```python
grouped_df = Unemployment_data2.groupby(["Region"])[" Estimated Unemployment Rate (%)"].aggregate("mean").reset_index()

grouped_df
# Mean Unemployment in each State
```

```python
plt.figure(figsize=(12,8))
sns.barplot(x=grouped_df['Region'], y=grouped_df[' Estimated Unemployment Rate (%)'])
plt.ylabel(' rate', fontsize=12)
plt.xlabel('Region', fontsize=12)
plt.title("Average of mean", fontsize=15)
plt.xticks(rotation='vertical')
plt.show()
```
